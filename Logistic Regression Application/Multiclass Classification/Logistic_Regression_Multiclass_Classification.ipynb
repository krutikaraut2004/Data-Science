Dataset Link : [click here!](https://scikit-learn.org/stable/auto_examples/datasets/plot_digits_last_image.html#:~:text=This%20dataset%20is%20made%20up,feature%20vector%20with%20length%2064.)

```python
# Import necessary libraries
import matplotlib.pyplot as plt  # Matplotlib for plotting
from sklearn.datasets import load_digits  # Scikit-learn for loading the digits dataset
```

```python
digits = load_digits()
```

```python
# Use the dir() function to list attributes and methods of the 'digits' object
dir(digits)
```

```python
# Access the data attribute of the 'digits' dataset and retrieve the first data point
digits.data[0]
```

```python
# Set the color map to grayscale for the Matplotlib plot
plt.gray()

# Display the image of the first data point in the 'digits' dataset
plt.matshow(digits.images[0])
```

```python
# Set the color map to grayscale for the Matplotlib plot
plt.gray()

# Display the images of the first five data points in the 'digits' dataset
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
# Access the target labels for the first five data points in the 'digits' dataset
digits.target[0:5]
```


**Create and train logistic regression model**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

**Measure accuracy of our model**

```python
model.score(X_test, y_test)
```

```python
# Display the image of the 68th data point in the 'digits' dataset
plt.matshow(digits.images[67])
```

```python
# Access the target label for the 68th data point in the 'digits' dataset
digits.target[67]
# ie what is the actual digit behind the image
```

```python
model.predict([digits.data[67]])
```

```python
# Use the trained model to make predictions on the features of the first five data points
model.predict(digits.data[0:5])
```

**Confusion Matrix**

```python
# Use the trained model to predict the target labels for the test set features
y_predicted = model.predict(X_test)
y_predicted
```

```python
# Import the confusion_matrix function from scikit-learn
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix using actual and predicted labels
cm = confusion_matrix(y_test, y_predicted)
cm


# y_test = truth
# Y_predicted = what the model predicts
```

```python
# Import necessary libraries
import seaborn as sn
import matplotlib.pyplot as plt

# Set the figure size for the heatmap
plt.figure(figsize=(10, 7))

# Create a heatmap using the confusion matrix with annotations
sn.heatmap(cm, annot=True)

# Set the labels for the x-axis and y-axis
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

The resulting heatmap visually represents the confusion matrix. The x-axis corresponds to the predicted labels, and the y-axis corresponds to the true (actual) labels. Each cell in the heatmap represents the count of instances where a true label (y-axis) was predicted as a certain class (x-axis). The annot=True parameter adds numerical annotations to each cell, displaying the counts.

This visualization helps to identify patterns in the model's predictions and errors. Brighter cells along the diagonal indicate correct predictions, while off-diagonal cells represent misclassifications.
